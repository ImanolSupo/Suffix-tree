```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar datos
df = pd.read_csv("benchmark_results_3.txt")

plt.rcParams['font.size'] = 14
plt.rcParams['axes.edgecolor'] = 'black'
plt.rcParams['axes.labelcolor'] = 'black'
plt.rcParams['xtick.color'] = 'black'
plt.rcParams['ytick.color'] = 'black'
plt.rcParams['text.color'] = 'black'
plt.rcParams['axes.grid'] = False  # sin cuadrcula

plt.figure(figsize=(10, 6))

plt.plot(df['n'], df['naive'], 
         marker='o', color='black', label='Naive', linewidth=1.2)

plt.plot(df['n'], df['mccreight'], 
         marker='s', color='black', label='McCreight', linewidth=1.2)

plt.plot(df['n'], df['ukkonen'], 
         marker='^', color='black', label='Ukkonen', linewidth=1.2)

plt.xlabel("Nmero de caracteres del texto")
plt.ylabel("Tiempo de construccin (ms)")
plt.title("Comparacin de Algoritmos de Construccin de Suffix Trees")


plt.legend()

plt.tight_layout()
plt.savefig("benchmark.png", dpi=1000)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar datos
df = pd.read_csv("benchmark_results_2.txt")

plt.rcParams['font.size'] = 14
plt.rcParams['axes.edgecolor'] = 'black'
plt.rcParams['axes.labelcolor'] = 'black'
plt.rcParams['xtick.color'] = 'black'
plt.rcParams['ytick.color'] = 'black'
plt.rcParams['text.color'] = 'black'
plt.rcParams['axes.grid'] = False  # sin cuadrcula
plt.figure(figsize=(10, 6))

plt.plot(df['n'], df['mccreight'],
         marker='s', color='black', label='McCreight', linewidth=1.3)

plt.plot(df['n'], df['ukkonen'],
         marker='^', color='black', label='Ukkonen', linewidth=1.3)

plt.xlabel("Nmero de caracteres del texto")
plt.ylabel("Tiempo de construccin (ms)")
plt.title("Comparacin McCreight vs Ukkonen en Construccin de Suffix Trees")
plt.legend()

plt.tight_layout()
plt.savefig("benchmark_mcc_vs_ukko.png", dpi=1000)
plt.show()
```
